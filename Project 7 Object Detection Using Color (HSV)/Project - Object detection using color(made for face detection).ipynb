```python
import cv2
import imutils
```

```python
meLower = (0,37,94)
meHigher = (60,255,255)
```

```python
cam = cv2.VideoCapture(0)
while True:
    flag, img = cam.read()
    #image preprocessing
    #img = imutils.resize(img, width = 600)
    blur = cv2.GaussianBlur(img, (11,11),0)
    hsv = cv2.cvtColor(blur, cv2.COLOR_BGR2HSV)
    
    #to extract only the need color object to minimise noise
    mask = cv2.inRange(hsv, meLower, meHigher)
    
    #to reduce the image by removing unnecessary gaps
    mask = cv2.erode(mask, None, iterations = 2)
    
    #to enlarge the now obtained img
    mask = cv2.dilate(mask, None, iterations = 2)
    
    #to find the borders over the most highlighted object of the now obtained img
    #we will later use it to find the area
    cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2]
    
    center = 0
    if len(cnts) > 0:
        #to find max contour area
        c = max(cnts, key = cv2.contourArea)
        
        #to find the location of center of circle which can enclose the object
        ((x,y),radius) = cv2.minEnclosingCircle(c)
        
        #or we can find center using area and moments
        M = cv2.moments(c)
        center = (int(M["m10"]/M["m00"]), int(M["m01"]/M["m00"]))
        
        if radius > 10:
            #draw a dynamic circle
            cv2.circle(img, (int(x),int(y)), int(radius), (0,255,255),2)
            
            #to make a cetroid dot in according to moments
            cv2.circle(img, center, 5, (0,0,255), -1)
            if radius > 250:
                print("stop")
            else:
                if(center[0]<150):
                    print("left")
                elif(center[0]>450):
                    print("right")
                elif(radius<250):
                    print("front")
                else:
                    print("stop")
    cv2.imshow("frame", img)
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break
cam.release()
cv2.destroyAllWindows()
        
```

```python
import cv2
import numpy
import imutils

#to initialise the camera
cam = cv2.VideoCapture(0)

#first frame to compare to next frame
firstFrame = None
#threshold for comparision
area = 1400

#to count images
count = 0
count2 = 0

#for continuous feed
while True:
    flag,img = cam.read()
    text = "Normal"
    img = imutils.resize(img, width = 500)
    grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gaussian = cv2.GaussianBlur(grayImg,(21,21),0)
    
    if firstFrame is None:
        firstFrame = gaussian
        continue
    
    #to find difference
    imgDiff = cv2.absdiff(firstFrame,gaussian)
    threshImg = cv2.threshold(imgDiff, 25 , 255, cv2.THRESH_BINARY)[1]
    
    #to remove the holes or dots in the img
    threshImg = cv2.dilate(threshImg,None, iterations = 2)
    
    #to make markers on the img
    cnts = cv2.findContours(threshImg.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts)
    
    
    #to make rectangle in the main img
    for c in cnts:
        if cv2.contourArea(c) < area:
            continue
        (x,y,w,h) = cv2.boundingRect(c)
        cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0),2)
        text = "moving object detected and total objects currently count "+str(count)
        text2 = "total objects till now "+str(count2)
        count+=1
        count2 += 1
        print(text2)
    cv2.putText(img, text, (10,20), cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0,0,255),2)
    cv2.putText(img, text2, (10,40), cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0,0,255),2)
    count = 0
    
    cv2.imshow("camFeed",img)
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break
cam.release()
cv2.destroyAllWindows()
```
